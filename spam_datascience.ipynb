Data Processing
-Lower case(converting the text to lower case)
-Tokenization(Breaking the sentences into words)
-Removing special characters()
-Removing stop words and punctuation
-Stemming(eg dancing will be converted to dance)

```python
import re #regular expressions(regex)
from nltk.tokenize import word_tokenize #module used to break text into indidvidual words or tokens
from nltk.corpus import stopwords #this imports 'stopwords' corpus from 'nltk' (e.g.'the','and','is')
from nltk.stem import PorterStemmer#converting running,dancing to dance,run
text="Go until jurong point, crazy.. Available only in bugis n great world la e buffet... Cine there got amore wat..."
text=text.lower()

#tokenization breaking the sentences into words
words=word_tokenize(text)

#Define regular expression pattern to remove special characters
pattern=r'[^a-zA-Z0-9]'

#remove special characters
words=[re.sub(pattern,'',word)for word in words]

#remove stop words and punctuation
stop_words = set(stopwords.words('english'))
words = [word for word in words if word not in stop_words and word.isalnum()]

# Stemming: Convert words to their root form
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in words]

# Print processed words
print("Processed words:", stemmed_words)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re

# Load your spam dataset
# Assuming you have a DataFrame named spam_train with a column 'text' containing the text data
# Example: spam_train = pd.read_csv('spam_dataset.csv')

# Perform text preprocessing
def preprocess_text(text):
    text = text.lower()
    words = word_tokenize(text)
    pattern = r'[^a-zA-Z0-9]'
    words = [re.sub(pattern, '', word) for word in words]
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words and word.isalnum()]
    stemmer = PorterStemmer()
    stemmed_words = [stemmer.stem(word) for word in words]
    return ' '.join(stemmed_words)

# Apply preprocessing to the text column of the DataFrame
spam_train['processed_text'] = spam_train['text'].apply(preprocess_text)

# Join the processed words into a single string
processed_text = ' '.join(spam_train['processed_text'])

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(processed_text)

# Display the word cloud using matplotlib
plt.figure(figsize=(9, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
spam_corpus=[]
for msg in spam_train[spam_train['target'] == 1]['text'].tolist():
    for word in msg:
        spam_corpus.append(word)
```

```python
len(spam_corpus)
```

```python
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming spam_corpus is a list containing words from the spam corpus
# If spam_corpus is a string containing all spam messages concatenated, you need to tokenize it first
# For example:
# spam_corpus = word_tokenize(spam_corpus_string)

# Create a Counter object to count word frequencies
word_counter = Counter(spam_corpus) #helps to count hashable objects

# Get the 30 most common words and their frequencies
most_common_words = word_counter.most_common(30)

# Create a DataFrame from the most common words
df_most_common = pd.DataFrame(most_common_words, columns=['Word', 'Frequency'])

# Create a bar plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Word', y='Frequency', data=df_most_common)
plt.xticks(rotation=90)  # Rotate x-axis labels to vertical
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.title('Top 30 Most Common Words in Spam Corpus')
plt.show()
```
